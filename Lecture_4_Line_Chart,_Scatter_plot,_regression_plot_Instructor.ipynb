# creating new dataframe with function 

```python
def straight_line(x,m,c):
  df = pd.DataFrame()
  df['x'] = x
  y = m*x + c
  df['y'] = y
  return df

```

#creating new straight line with function

```python
SL_1 = straight_line(np.arange(-10,11,1),3,1)
SL_2 = straight_line(np.arange(-10,11,1),-1,1)
SL_3 = straight_line(np.arange(-10,11,1),4,-1)
SL_4 = straight_line(np.arange(-10,11,1),-1,-4)
SL_5 = straight_line(np.arange(-10,11,1),1,0)
SL_6 = straight_line(np.arange(-10,11,1),-1,-1)
```

# plot function

```python
def plot_stl(x_axis, y_axis, title):
  plt.plot(x_axis, y_axis, marker='.', color='red')
  plt.vlines(0, -15, 15, linestyles=':', colors="black")
  plt.hlines(0, -10, 10, linestyles=':', colors="black")
  plt.xticks(np.arange(-10,11,5))
  plt.yticks(np.arange(-15,16,5))
  plt.title(title)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.tight_layout()
```

#plotting

```python
plt.figure(figsize=(20,12))



plt.subplot(2,3,1)
plot_stl(SL_1['x'],SL_1['y'], 'y = mx+c where m = 3 & c=1')

plt.subplot(2,3,2)
plot_stl(SL_2['x'],SL_2['y'], 'y = mx+c where m = -1 & c=1')

plt.subplot(2,3,3)
plot_stl(SL_3['x'],SL_3['y'], 'y = mx+c where m = 4 & c= -1')

plt.subplot(2,3,4)
plot_stl(SL_4['x'],SL_4['y'], 'y = mx+c where m = -1 & c= -4')

plt.subplot(2,3,5)
plot_stl(SL_5['x'],SL_5['y'], 'y = mx+c where m = 1 & c= 0')

plt.subplot(2,3,6)
plot_stl(SL_6['x'],SL_6['y'], 'y = mx+c where m = -1 & c= -1')
```
